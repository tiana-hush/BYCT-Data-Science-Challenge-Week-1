```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define dimensions(Height and width) of the image
height = 400
width = 800

# Generate a random grayscale image, including my dimensions as size
Grayscale_img = np.random.randint(0, 255, size=(height, width))

# Displays the image using imshow with 'magma' colormap
plt.imshow(Grayscale_img, cmap='magma')
plt.title('Random Grayscale Image')
plt.show()
```

```python
def generate_random_grayscale(height, width, min_intensity=0, max_intensity=255): # Generate random integers in the specified intensity range
    
    Grayscale_img = np.random.randint(min_intensity, max_intensity + 1, size=(height, width))
    return Grayscale_img

def display_grayscale_image(image, title='Random Grayscale Image', cmap='magma'):
    plt.imshow(image, cmap=cmap)
    plt.colorbar()
    plt.title(title)
    plt.show()

# Define dimensions(Height and width) of the image
height = 400
width = 800

# Define the intensity range for grayscale values
min_intensity = 0
max_intensity = 127  

# Generate images with different values within the specified intensity range(e.g., 0-127) 
Grayscale_img = generate_random_grayscale(height, width, min_intensity, max_intensity)

# Display the image using imshow with 'magma' colormap
display_grayscale_image(Grayscale_img, f'Random Grayscale Image ({min_intensity}-{max_intensity})')
```
